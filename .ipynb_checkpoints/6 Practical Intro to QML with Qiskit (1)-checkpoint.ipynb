# QML Example: Variational Quantum Classifier

4 things we need to build one of these:
1. A data encoding strategy: for simplicity we use a Qiskit Feature Map
2. A parameterised variational ansatz: the processing layers of our **QNN**
3. An optimiser: here we use Cobyla (has been shown to work well for VQCs)
4. An output transformation: turning our qubit measurements into a probability distribution for class label prediction

From the classical example, we have the Iris dataset in the following format:
- Normalised into $\{0,1\}$ via the *MinMaxScaler*
- Labels have been seperated from the main dataframe: *train_labels* and *test_labels*
- We have an 80/20 random holdout split (note that it is not stratified, but as iris is uniformly distributed in the class label, this doesn't really matter) 

For example:

```python
df = pd.DataFrame(features, columns=iris_data.feature_names)
df["class"] = pd.Series(iris_data.target)
df.head()
```

# Encoding: Feature Map

We want to build a Feature Map on the basis of a few decisions:
1. Do we all or only some of the features?
2. Do we want to repeat any of the features in the map?
3. Are we happy to relinquish control over how the encoding is performed?

Note, that this manner of Feature Map will:
- Place each feature on its own qubit wire
- Use entanglement ($CX$ gates) to "share" information between qubits
- Will (in this case) use multi-parameter gates to generate something akin to a feature space -- note that the data features are loaded multiple times for this to work

```python
num_features = features.shape[1]

feature_map = ZZFeatureMap(feature_dimension=num_features, reps=1)
feature_map.decompose().draw(output="mpl", style="clifford", fold=20)
```

# Defining the QML (i.e. VQC) model

In principle, we need to make a few key decisions here (and we will expand upon these in future lectures):
1. how many qubits will we use? 
2. how many **reps** (i.e. repetitions) of the ansatz do we want? 
3. which optimiser are we going to use?
4. how much time (here iterations) will we give the optimiser?

## How many qubits?

This is in part answered by the decision to use the type of Feature Map we picked previously: it assigned 1 qubit per feature.

We wouldn't have had to do this, and we'll see some options for this later in the lecture.

As a general comment, the power of quantum is the ability to scale in the number of qubits, but the challenge is using more qubits effectively:
- more qubits means more entanglement
- more entanglement means increased circuit depth
- increased circuit depth means needing better strategies to deal with errors and noise

Similarly, the computational complexity of simulating the quantum systems scales quadratically with the number of qubits. 

There are hard limits to what your laptop can do. Modern day laptops can handle the simlation of only 10-15 qubit systems.

## How many reps?

- Each *rep* is like a layer in a (Q)NN
- As we increase the number of layers, we also increased the propensity for NISQ-like errors to compound and cause problems. 
- In an error-free setting (such as the state vector simulator) we can easily fool ourselves into thinking more layers means better
- In a NISQ setting, we want to minimise circuit depth: noise (from error) is bad, but we can tolerate some noise! 
- More reps means more parameters (here, basically model weights) to optimise -- these weights are different to the feature parameters from the Feature Map

## Optimiser

- There are many options, we'll explore some of these and compare their performance
- Each option has its own parameters (i.e. effectively hyperparameters) to consider
- There are progressively more and more options being developed
- In the early days (so 2020/1) optimisers were more general purpose, but recently we are seeing optimisers actively targetting the use case of VQML circuits.

# Defining our ansatz

```python
ansatz = RealAmplitudes(num_qubits=num_features, reps=3)
ansatz.decompose().draw(output="mpl", style="clifford", fold=20)
```

Repeating 3 times, means 16 parameters (or model weights).

This subcircuit, will be concatenated with our Feature Map

# Define the optimiser

```python
optimizer = COBYLA(maxiter=100)
sampler = Sampler()

#array of loss (or similar) values for each iteration
objective_func_vals = []
plt.rcParams["figure.figsize"] = (12, 6)
```

100 iterations, hard to know a good value for this ex ante: need to experiment

Could be a case to consider **standard** ML tricks like early stopping here

The *Sampler* is a state vector simulator, and can be omitted if running in simulation (as will often be the case). If omitted, the VQC constructor will make an instance of it for us.

Qiskit gives the option for a call back function to keep track on how training is going, we can plot this in a manner similar what we are used to in deep learning. 

This gives us a sense of how the model is converging in terms of its error (loss).

You can define this function to do whatever you like, but it must have the signature as defined below, and take the inputs as listed.

```python
def callback_graph(weights, obj_func_eval):
    clear_output(wait=True)
    objective_func_vals.append(obj_func_eval)
    plt.title("Objective function value against iteration")
    plt.xlabel("Iteration")
    plt.ylabel("Objective function value")
    plt.plot(range(len(objective_func_vals)), objective_func_vals)
    plt.show()
```

## Constructing the VQC

```python
vqc = VQC(
    sampler=sampler,
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=optimizer,
    callback=callback_graph,
)

# clear objective value history
objective_func_vals = []
```

```python
start = time.time()
vqc.fit(train_features, train_labels)
elapsed = time.time() - start
```

```python
train_score_q4 = vqc.score(train_features, train_labels)
test_score_q4 = vqc.score(test_features, test_labels)

print(f"Training time: {round(elapsed)} seconds")
print(f"Quantum VQC on the training dataset: {train_score_q4:.2f}")
print(f"Quantum VQC on the test dataset:     {test_score_q4:.2f}")
```

```python
def generate_train_test(features, labels, nPCs):

    famd = prince.FAMD(
        n_components=nPCs,
        copy=True,
        check_input=True,
        random_state=algorithm_globals.random_seed,
        engine="sklearn",
        handle_unknown="error"
    )

    train_features_all, test_features_all, train_labels_all, test_labels_all = train_test_split(
        features, labels, train_size=0.8, random_state=algorithm_globals.random_seed
    )

    train_labels_all = train_labels_num.astype('string')
    train_labels_all = train_labels_all.array
    test_labels_all = test_labels_num.astype('string')
    test_labels_all = test_labels_all.array

    train_features_all.reset_index(drop=True, inplace=True)
    test_features_all.reset_index(drop=True, inplace=True)

    train_famd = famd.fit(train_features_all)

    train_famd_X = train_famd.row_coordinates(train_features_all)
    test_famd_X = train_famd.row_coordinates(test_features_all)

    return train_famd_X, train_labels_all, test_famd_X, test_labels_all
```

```python
from qiskit.circuit.library import EfficientSU2
from qiskit.circuit.library import ZFeatureMap

num_features = 7
train, labels_train, test, labels_test = generate_train_test(features_new, labels, num_features)

feature_map = ZFeatureMap(feature_dimension=num_features, reps=2)

ansatz = EfficientSU2(num_qubits=num_features, reps=1)
optimizer = COBYLA(maxiter=100)
sampler = Sampler()

qc = feature_map & ansatz
qc.decompose().draw(output="mpl", style="clifford", fold=20)
```

```python
def train_vqc(train, labels_train, test, labels_test, sampler, feature_map, ansatz, optimizer):
    vqc = VQC(
        sampler=sampler,
        feature_map=feature_map,
        ansatz=ansatz,
        optimizer=optimizer,
        callback=callback_graph
    )

    start = time.time()
    vqc.fit(train, labels_train)
    elapsed = time.time() - start

    train_score_q4 = vqc.score(train, labels_train)
    test_score_q4 = vqc.score(test, labels_test)

    print(f"Training time: {round(elapsed)} seconds")
    print(f"Quantum VQC on the training dataset: {train_score_q4:.2f}")
    print(f"Quantum VQC on the test dataset:     {test_score_q4:.2f}")
```

```python
# clear objective value history
objective_func_vals = []

train_vqc(train, labels_train, test, labels_test, sampler, feature_map, ansatz, optimizer)
```
